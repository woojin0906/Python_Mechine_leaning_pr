```python
import matplotlib.pyplot as plt
import numpy as np

x = 2 * np.random.rand(100, 1) # 0~1 사이의 균일분포 -> shape : 100 x 1
y = 4 + 3*x + np.random.randn(100, 1) # normal 분포 mu=0, var=1 shape : 100x1
plt.scatter(x, y)
plt.show()
```

```python
x_bias = np.c_[np.ones((100,1)), x] # 배열 형태의 새로운 열을 추가함

theta_best = np.linalg.inv(x_bias.T.dot(x_bias)).dot(x_bias.T).dot(y)
theta_best
```

```python
x_new = np.array([[0], [2]])
x_new_b = np.c_[np.ones((2,1)), x_new]\
    
prediction = x_new_b.dot(theta_best)
prediction

plt.plot(x_new, prediction, "r-")
plt.plot(x,y,"b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
print(lin_reg.intercept_, lin_reg.coef_)
```
